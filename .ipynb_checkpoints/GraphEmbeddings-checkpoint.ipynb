```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data/recipes.csv")
amount_column = pd.DataFrame(df).shape[0]
amount_line = pd.DataFrame(df).shape[1]
print(df.head(10))
print(f'Общее количество колонок в таблице: {amount_column}')
print(f'Общее количество cтрок в таблице: {amount_line}\n')

print("Исходные пропущенные значения:\n", df.isna().sum())
```

```python
test_size = 0.4
random_state = 42

df["train"] = (np.random.rand(len(df)) < 0.6)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]  # целевая функция

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

print("Размер обучающей выборки:", X_train.shape)
print("Размер тестовой выборки:", X_test.shape)
```

Создаем граф знаний на основе информации
Cгенерируем новое представление набора данных, где каждая точка данных представляет собой тройку в виде:

<subject, predicate, object>
А также создадим сущности ( субъекты и объекты)

```python
df["drink_id"] = "Drink " + df["drink_type"]
df["milk_id"] = df["milk (in cl)"].astype(str).apply(lambda x: "Milk " + x)
df["ice_cubes_id"] = df["ice cubes"].astype(str).apply(lambda x: "IceCubes" + x)
df["size_id"] = "Size" + df["size (mixed drinks)"]
df["coffee_dose_ristrettos_id"] = df["coffee dose (in number of ristrettos)"].fillna(0).astype(str).apply(lambda x: "CoffeeDoseRistrettos" + x)
df["coffee_dose_espressos_id"] = df["coffee dose (in number of espressos)"].fillna(0).astype(str).apply(lambda x: "CoffeeDoseEspressos" + x)
df["aroma_pumps_id"] = df["aroma pumps (=10 ml)"].fillna(0).astype(str).apply(lambda x: "AromaPumps" + x)
```

Создаем триплеты

```python
triples = []

for _, row in df.iterrows():
    # Напиток и его характеристики
    drink = (row["drink_id"], "hasMilkAmount", row["milk_id"])
    ice_cubes = (row["drink_id"], "hasIceCubesAmount", row["ice_cubes_id"])
    size = (row["drink_id"], "hasSize", row["size_id"])
    aroma_pumps = (row["drink_id"], "hasAromaPumps", row["aroma_pumps_id"])

    # Количество кофе (если указано)
    if pd.notnull(row["coffee dose (in number of ristrettos)"]):
        coffee_dose = (row["drink_id"], "hasCoffeeDoseInEspressos", row["coffee dose (in number of ristrettos)"])
        coffee_dose_espressos = (row["drink_id"], "hasCoffeeDoseInEspressos", row["coffee dose (in number of espressos)"])
    else:
        coffee_dose = (row["drink_id"], "hasCoffeeDoseInRistrettos", "None")
        coffee_dose_espressos = (row["drink_id"], "hasCoffeeDoseInEspressos", "None")

    triples.extend([drink, ice_cubes, size, coffee_dose, coffee_dose_espressos])

for triple in triples:
    print(triple)
```

```python
triples_df = pd.DataFrame(triples, columns=["subject", "predicate", "object"])
filtered_triples = triples_df[(triples_df.subject == "Drink ristretto") | (triples_df.object == "Drink ristretto")]
print(filtered_triples)
```

```python
from ampligraph.evaluation import train_test_split_no_unseen
import numpy as np

# Разделяем данные на обучающую и тестовую выборку
X_train, X_valid = train_test_split_no_unseen(np.array(triples), test_size=10)

print('Train set size: ', X_train.shape)
print('Test set size: ', X_valid.shape)
```

```python
from ampligraph.latent_features import ScoringBasedEmbeddingModel
from ampligraph.latent_features.loss_functions import get as get_loss
from ampligraph.latent_features.regularizers import get as get_regularizer

model = ScoringBasedEmbeddingModel(k=100,
                                   eta=20,
                                   scoring_type='ComplEx',
                                   seed=0)

# Optimizer, loss and regularizer definition
optimizer = 'adam'
loss = get_loss('multiclass_nll')
regularizer = get_regularizer('LP', {'p': 3, 'lambda': 1e-5})
print(tf.__version__)
# Compilation of the model
model.compile(optimizer=optimizer, loss=loss, entity_relation_regularizer=regularizer)
```
